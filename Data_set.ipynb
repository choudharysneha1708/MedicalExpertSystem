```python
import pandas as pd
# Load the dataset
data = pd.read_csv('Data.csv')
data = data.drop('Target Organ', axis=1)
# Iterate through each row in the dataset
for index, row in data.iterrows():
    disease_name = row['Diseases']  # Get disease name from the 'Disease' column
    print(f"Disease Name: {disease_name}")

    # Iterate through symptoms columns (starting from the second column)
    for symptom, value in row.items():
        if symptom != 'Disease' and value in [1, 2, 3]:
            print(f"{symptom}: Value {value}")

    print()  # Add an empty line for better readability
```
