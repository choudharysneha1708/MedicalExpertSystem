```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
import pandas as pd

# Load your dataset into a pandas DataFrame
data = pd.read_csv('data.csv')

# Drop the "Target Organ" column
data = data.drop('Target Organ', axis=1)

# Create a dictionary to store symptoms associated with each disease
symptoms_dict = {}

for index, row in data.iterrows():
    disease = row['Diseases']
    symptoms = [column for column in data.columns if row[column] == 1]  # Assuming 1 indicates the presence of a symptom
    if disease in symptoms_dict:
        symptoms_dict[disease].extend(symptoms)
    else:
        symptoms_dict[disease] = symptoms
# Print the symptoms for each disease
for disease, symptoms in symptoms_dict.items():
    print(f"Disease: {disease}")
    symptoms.sort()
    i = 1
    for symptom in symptoms:
      print('\t ', i ,'# ', symptom)
      i += 1
    print("\n")
```

```python
data = pd.read_csv('data.csv')

# Extract the disease column
disease_column = 'Diseases'  # Replace with the actual column name
diseases = data[disease_column]

symptoms_columns = data.columns[2:]

# Calculate the number of symptoms marked as 1 for each disease
symptoms_count = data[symptoms_columns].sum(axis=1)

# Create a bar chart
plt.figure(figsize=(30, 15))
plt.bar(diseases, symptoms_count, color='skyblue')
plt.xlabel('Disease')
plt.ylabel('Number of Symptoms (Marked as 1)')
plt.title('Number of Symptoms Marked as 1 by Disease')
plt.xticks(rotation=90)  # Rotate the disease labels for better readability

# Display the plot
plt.tight_layout()
plt.show()
plt.savefig('bar-graph(diseases_X_frequency).png')
```

```python
target_organ_columns = [col for col in data.columns if set(data[col].unique()) == {0, 1}]

# Initialize an empty dictionary to store the combined counts for all target columns
combined_counts = {}

# Iterate through the target organ columns and aggregate the counts
for target_column in target_organ_columns:
    counts = data[data[target_column] == 1][target_column].count()
    combined_counts[target_column] = counts

# Create a pie chart with adjusted label font size and positioning
plt.figure(figsize=(8, 8))
plt.pie(
    combined_counts.values(),
    labels=combined_counts.keys(),
    autopct='%1.1f%%',
    startangle=140,
    textprops={'fontsize': 5}  # Adjust the font size as needed
)
plt.title('Distribution of Diseases or Symptoms by Target Organ')

# Display the plot
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular
plt.show()
plt.savefig('pie-chart(diseases_X_symptoms).png')
```

```python
import networkx as nx

# Load data from the CSV file (replace 'your_data.csv' with the actual file path)
data = pd.read_csv('data.csv')

# Create an empty graph
G = nx.Graph()

# Extract unique target organs, diseases, and symptoms
unique_target_organs = data['Target Organ'].unique()
unique_diseases = data['Diseases'].unique()
unique_symptoms = data.columns[2:]  # Assuming the symptom columns start from the third column

# Add nodes for target organs, diseases, and symptoms
G.add_nodes_from(unique_target_organs, node_type='Target Organ')
G.add_nodes_from(unique_diseases, node_type='Disease')
G.add_nodes_from(unique_symptoms, node_type='Symptom')

# Add edges to connect target organs to diseases and diseases to symptoms
for _, row in data.iterrows():
    target_organ = row['Target Organ']
    disease = row['Diseases']
    symptoms = [symptom for symptom, value in row[2:].items() if value == 1]

    # Connect target organs to diseases
    G.add_edge(target_organ, disease, relation='Target Organ to Disease')

    # Connect diseases to symptoms
    for symptom in symptoms:
        G.add_edge(disease, symptom, relation='Disease to Symptom')

# Create a more organized layout to reduce overlapping
pos = nx.spring_layout(G, seed=42, k=0.25)  # Adjust 'k' to control node separation

# Draw the network graph
plt.figure(figsize=(30, 20))
node_size = 3000
nx.draw(G, pos, with_labels=True, node_size=node_size, node_color='lightblue', font_size=8, font_color='black')

# Draw edge labels to indicate the type of relation
edge_labels = nx.get_edge_attributes(G, 'relation')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8, font_color='red')

plt.title('Network Graph of Target Organs, Diseases, and Symptoms')
plt.savefig('network-graph(diseases_to_symptoms).png')
plt.axis('off')
plt.show()
```
